# FEATURE SELECTION USING VARIANCE THRESHOLD
AIM: DROPPING LOW VARIANCE FEATURES

In this step we would use VarianceThreshold to removes features with a variance less than the specified threshold.

```python
import pandas as pd
import numpy as np
# importing MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
# importing train_test_split
from sklearn.model_selection import train_test_split
# importing VarianceThreshold
from sklearn.feature_selection import VarianceThreshold
# importing warnings
import warnings
```

```python
pd.pandas.set_option('display.max_columns',None)
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('project_data.csv')
```

```python
df.info()
```

Since there are no null values, we can easily proceed with feature selection

```python
# Droping date column
df.drop(labels = ['Date'], axis = 1, inplace = True)
```

```python
df
```

```python
# normalize the dataset using the MinMaxScaler
scaler = MinMaxScaler()
df2 = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
# taking Adj Close  as target variable
# independent variables x
X = df2.drop(columns=['Adj Close'])
# dependent variable y
y = df2['Adj Close']
```

```python
# using a 70-30 train-test set split with a random state of 42
# this separation helps to prevent overfitting
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size=0.3, random_state=42)
```

```python
# shape of test and train data
X_train.shape, X_test.shape
```

```python
# applying the variance threshold for features with a variance higher than 0.1
var_thres = VarianceThreshold(threshold = 0.1)
# Note: fit would be applied only on x_train while x_test will only undergo transformation
var_thres.fit(X_train)
```

```python
var_thres.fit_transform(X_train)
```

```python
# total number of features with a variance higher than 0.1
sum(var_thres.get_support())
```

```python
# low variance columns
low_var_columns = [column for column in X_train
                  if column not in X_train.columns[var_thres.get_support()]]
low_var_columns
```

```python
len(low_var_columns)
```

Hence there are 14 features with variance less than 0.1

```python
# dropping low variance columns
X_train = X_train.drop(low_var_columns, axis = 1).reset_index(drop = True)
X_train
```

```python
# tansforming X_test
X_test = X_test.drop(low_var_columns, axis = 1).reset_index(drop = True)
X_test
```
